```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
import json

result = []
for _, row in df.iterrows():
    entry = {
        "instruction": f"target_Tg={row['Tg']}",
        "output": row['PSMILES']
    }
    result.append(entry)

with open('Tg_data.json', 'w', encoding='utf-8') as f:
    for item in result:
        f.write(json.dumps(item, ensure_ascii=False) + '\n')
```

```python
import json
import random

# 读取Tg_data.json
with open('Tg_data.json', 'r', encoding='utf-8') as f:
    data = [json.loads(line) for line in f]

# 打乱数据
random.shuffle(data)

# 按照0.2比例分割
split_idx = int(len(data) * 0.8)
train_data = data[:split_idx]
val_data = data[split_idx:]

# 写入train.json
with open('train.json', 'w', encoding='utf-8') as f_train:
    for item in train_data:
        f_train.write(json.dumps(item, ensure_ascii=False) + '\n')

# 写入val.json
with open('val.json', 'w', encoding='utf-8') as f_val:
    for item in val_data:
        f_val.write(json.dumps(item, ensure_ascii=False) + '\n')
```
